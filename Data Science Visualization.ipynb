```python
#Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("retail_sales_dataset.csv")

#To Display First 5 Rows
print("First 5 Rows:")
print(df.head())

#For Check missing values
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
# For Bar Chart
plt.figure(figsize=(8, 6))
sns.countplot(x='Gender', data=df)
plt.title('Bar Chart - Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
#For Pie Chart
category_count = df['Product Category'].value_counts()

labels = category_count.index
plt.figure(figsize=(6, 6))
plt.pie(category_count, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Pie Chart - Product Category Distribution')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('retail_sales_dataset.csv')  
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Gender', hue='Product Category')

plt.title('Gender-wise Product Category Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.legend(title='Product Category')
plt.tight_layout()
plt.show()
```

```python
# For Histogram
plt.figure(figsize=(8, 6))
plt.hist(df['Age'], bins=20, color='skyblue', edgecolor='black')
plt.title('Histogram - Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count') 
plt.show()
```

```python

df['AgeGroup'] = pd.cut(df['Age'], bins=[15,25,35,45,55,65], labels=['15-25','26-35','36-45','46-55','56-65'])
grouped = df.groupby(['AgeGroup', 'Product Category'], observed=True).size().reset_index(name='Count')
pivot_df = grouped.pivot(index='AgeGroup', columns='Product Category', values='Count').fillna(0)

pivot_df.plot(kind='line', marker='o', figsize=(10, 6))
plt.title('Product Category Preference by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Number of Customers')
plt.grid(True)
plt.legend(title='Product Category')
plt.tight_layout()
plt.show()

```
